### - Read the data into a data frame.

```python
import pandas as pd
df = pd.read_csv('pitching.csv')
df
```

### - Display all pitchers in 2015 under each team sorted by their last names.

```python
df_2015 = df.loc[df['yearID'] == 2015]
unique_team = pd.unique(df_2015['teamID'])

df_2015_lastname_team = df_2015[['teamID', 'playerID', 'nameFirst', 'nameLast', 'nameGiven']].groupby(by = 'teamID')


for i in range(len(unique_team)):
    team_i = df_2015_lastname_team.get_group(unique_team[i])
    team_i = team_i.sort_values(by = 'nameLast')
    print('\n[', unique_team[i], ']\n')
    print(team_i)
```

### – Display all pitchers in 2015 sorted by their ERAs. This allows you to identify the best and worst pitchers.

```python
df_2015_ERA = df_2015[['playerID', 'ERA']].sort_values(by = 'ERA')
df_2015_ERA
```

### – Display all pitchers in 2015 under each team sorted by their ERAs.

```python
df_2015_ERA_team = df[['teamID', 'playerID', 'ERA']]
unique_team = pd.unique(df_2015_ERA_team['teamID'])
df_2015_ERA_team = df_2015_ERA_team.groupby(by = 'teamID')

for i in range(len(unique_team)):
    team_i = df_2015_ERA_team.get_group(unique_team[i])
    team_i = team_i.sort_values(by = 'ERA')
    print('\n[', unique_team[i], ']\n')
    print(team_i)
```

### – Display all pitchers in 2015 under each team sorted by their strike outs (SO).

```python
df_2015_SO = df_2015[['playerID', 'teamID', 'SO']].sort_values(by = 'SO')
unique_team = pd.unique(df_2015_SO['teamID'])
df_2015_SO = df_2015_SO.groupby(by = 'teamID')

for i in range(len(unique_team)):
    team_i = df_2015_SO.get_group(unique_team[i])
    team_i = team_i.sort_values(by = 'SO')
    print('\n[', unique_team[i], ']\n')
    print(team_i)
```

### – Calculate the average ERA for each team, list the teams by the average ERA in a descending order.

```python
df_ERA = df[['teamID', 'ERA']]
df_ERA = df_ERA.groupby(by = 'teamID').mean().sort_values(by = 'ERA', ascending = False)
df_ERA
```

### – Calculate the total number of home runs allowed (HRs) for each team, and list the team by the count in an ascending order.

```python
df_HR = df[['teamID', 'HR']]
df_HR = df_HR.groupby(by = 'teamID').sum(level = 'teamID').sort_values(by = 'HR')
df_HR
```

### – Calculate each team’s average ERA from 2011 to 2015, and list the result as a table, where each column is a year, each row is a team.

```python
df_ERA = df[['teamID', 'ERA', 'yearID']]
df_ERA_year = df_ERA.groupby(by = 'yearID')

for i in range(2011, 2016, 1):
    df_ERA_i = df_ERA_year.get_group(i)
    df_ERA_i_mean = df_ERA_i[['teamID', 'ERA']].groupby(by = 'teamID').mean()
    df_ERA_i_mean.columns = [i]
    if i == 2011:
        df_ERA = df_ERA_i_mean
    else:
        df_ERA = pd.concat([df_ERA, df_ERA_i_mean], axis = 1, sort = False)

df_ERA    
```

### – Calculate the average ERAs for American League (AL) and National League (NL) from 2011 to 2015, and list the result as a table, where each column is a league and each row is a year.

```python
# AL
df_lgID_AL = df[['lgID', 'ERA', 'yearID']].groupby(by = 'lgID').get_group('AL')
df_AL_ERA_mean = df_lgID_AL[['ERA', 'yearID']].groupby(by = 'yearID')['ERA'].mean()

# NL
df_lgID_NL = df[['lgID', 'ERA', 'yearID']].groupby(by = 'lgID').get_group('NL')
df_NL_ERA_mean = df_lgID_NL[['ERA', 'yearID']].groupby(by = 'yearID')['ERA'].mean()

# Concate AL and NL
df_AL_NL_EAR_mean = pd.concat([df_AL_ERA_mean, df_NL_ERA_mean], axis = 1, sort = False)
df_AL_NL_EAR_mean.columns = ['AL', 'NL']

df_AL_NL_EAR_mean = df_AL_NL_EAR_mean.loc[[2011, 2012, 2013, 2014, 2015]]
df_AL_NL_EAR_mean
```

### – List the pitchers who had the most wins in each of the years from 2011 to 2015.

```python
df_pitcher = df[['yearID', 'playerID', 'W']]
df_pitcher_year = df_pitcher.groupby(by = 'yearID')

for i in range(2011, 2016, 1):
    df_pitcher_i = df_pitcher_year.get_group(i)
    pitcher_i_w = df_pitcher_i[['yearID', 'playerID', 'W']].sort_values(by = 'W', ascending = False).head(1)
   
    if i == 2011:
        df_pitcher = pitcher_i_w
    else:
        df_pitcher= pd.concat([df_pitcher, pitcher_i_w], axis = 0, ignore_index=True)

df_pitcher
```
